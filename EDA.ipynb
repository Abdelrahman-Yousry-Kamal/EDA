```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('movies_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
avg_revenue = df.groupby('year')['revenue_million'].mean()

plt.figure(figsize=(10,5))
plt.plot(avg_revenue.index, avg_revenue.values, marker='o')
plt.title("Average Movie Revenue per Year")
plt.xlabel("year")
plt.ylabel("Average Revenue (Million $)")
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,6))
axes[0].scatter(df['rating'], df['votes'], alpha=0.6, color="purple")
axes[0].set_title("Rating vs Votes")
axes[0].set_xlabel("Rating")
axes[0].set_ylabel("Votes")
axes[1].scatter(df['budget_million'], df['revenue_million'], alpha=0.6, color="orange")
axes[1].set_title("Budget vs Revenue")
axes[1].set_xlabel("Budget (Million $)")
axes[1].set_ylabel("Revenue (Million $)")

plt.tight_layout()
plt.show()
```

```python
movies_per_year = df['year'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.bar(movies_per_year.index, movies_per_year.values, color="skyblue")
plt.title("Number of Movies Released per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.shape
```

```python
penguins = penguins.dropna(subset=["body_mass_g"])
```

```python
plt.figure(figsize=(8,6))
plt.hist(penguins["body_mass_g"], bins=20, color="teal", edgecolor="black", alpha=0.7)
plt.title("Distribution of Penguin Body Mass")
plt.xlabel("Body Mass (g)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g", palette="Set1")
plt.title("Penguin Body Mass Across Species")
plt.show()
```

```python
sns.violinplot(data=penguins, x="species", y="flipper_length_mm", palette="Set1")
plt.title("Flipper Length Across Penguin Species")
plt.show()
```

```python
sns.histplot(data=df, x="rating", bins=10, kde=True, color="skyblue")
plt.title("Distribution of Movie Ratings")
plt.show()
```

```python
sns.barplot(data=df, x="genre", y="revenue_million", estimator="mean", palette="Set1")
plt.title("Average Revenue by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x="genre", palette="Set1")
plt.title("Number of Movies by Genre")
plt.show()
```

```python
sns.pairplot(penguins.dropna(), diag_kind="kde")
plt.suptitle("Penguins Pairplot (Numeric Features)", y=1.02)
plt.show()
```

```python
corr = df[['budget_million', 'revenue_million', 'rating', 'votes']].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.regplot(data=df, x="budget_million", y="revenue_million")
plt.title("Regression: Movie Budget vs Revenue")
plt.xlabel("Budget (Million $)")
plt.ylabel("Revenue (Million $)")
plt.show()
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.suptitle("Penguin Flipper Length vs Body Mass")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.boxplot( data=penguins, x="species", y="body_mass_g", palette="coolwarm")
plt.title("Penguin Body Mass Across Species (Styled)")
plt.xlabel("Penguin Species")
plt.ylabel("Body Mass (g)")
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.histplot(data=df, x="rating",kde=True, color="green")
plt.title("Distribution of Movie Ratings")
plt.xlabel("Movie Rating")
plt.ylabel("Number of Movies")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
sns.lmplot(
    data=df,
    x="budget_million",
    y="rating",
    scatter_kws={"alpha":0.6},
    line_kws={"color":"red"}
)
plt.title("Do Bigger Budgets Lead to Higher Ratings?")
plt.xlabel("Budget (Million $)")
plt.ylabel("Rating")
plt.show()
```

```python
sns.pairplot(
    penguins,
    hue="species",
    diag_kind="kde",
    palette="Set1"
)
plt.suptitle("Penguin Species Comparison Across Measurements", y=1.02)
plt.show()
```
